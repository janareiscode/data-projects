```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('dance_genre.csv', sep = ';', decimal=',')

# –ü—Ä–æ—Å–º–æ—Ç—Ä –ø–µ—Ä–≤—ã—Ö –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Å—Ç—Ä–æ–∫ —Ç–∞–±–ª–∏—Ü—ã
df.head()
```

```python

df.columns = ['Time', 'Old', 'Sex', 'Years_of_dancing', 'Style', 'Fav_DanceTeam', 'Fav_series', 'Fav_movie', 'Fav_anime', 'GPA', 'Happy', 'YU_Dance', 'Fell_better', 'Frequency', '', '']
df
```

```python
gr = df.groupby(['Style'])['GPA'].agg('mean').round(2)
gr
```

```python
no_dance = df[df['Style'] == "Don't dance"]['GPA'].mean()
no_dance
```

```python
dance = df[df['Style'] != "Don't dance"]['GPA'].mean()
dance
```

```python
gr = df.groupby(['Style'])['GPA'].agg('mean').round(2)
colors = ['skyblue' if style != "Don't dance" else 'lightcoral' for style in gr.index]
         
# –°–æ–∑–¥–∞–Ω–∏–µ –±–∞—Ä-–ø–ª–æ—Ç–∞

plt.figure(figsize=(10, 4))
plt.barh(gr.index, gr.values, color=colors)  # –ü–æ–º–µ–Ω—è–ª–∏ bar –Ω–∞ barh –¥–ª—è –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–≥–æ –±–∞—Ä-–ø–ª–æ—Ç–∞
plt.xlabel('Average GPA')
plt.xlim(1.5, 3.7)
plt.ylabel('Style of Dance')
plt.title('Average GPA for Various Styles of Dance')
plt.grid(axis='x', linestyle='--', alpha=0.7)  # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å–µ—Ç–∫–∏
plt.tight_layout()  # –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∞—è –ø–æ–¥—Å—Ç—Ä–æ–π–∫–∞ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤ –≥—Ä–∞—Ñ–∏–∫–∞ –¥–ª—è –ª—É—á—à–µ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è
plt.savefig('styleGPA')
plt.show()
```

Non-dancers have higher GPA than dancers.
The highest GPA have Contemp and Afro dancers(GPA>3.9) more than non-dancers.
The lowest GPA gave Jazz-funk group(GPA = 2.65)

```python
grouped = df.groupby(["Sex", "Style"]).size().unstack(fill_value=0)

# –°—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
grouped.plot(kind="bar", stacked=True)
plt.title("Distribution of dance styles by gender")
plt.xlabel("Sex")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.legend(title="Style", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('style_gender.png')
plt.show()
```

```python
filtered_df = df[~df['Fav_anime'].isin(["Don't watch"])]

# –°–æ–∑–¥–∞–µ–º —Å–≤–æ–¥–Ω—É—é —Ç–∞–±–ª–∏—Ü—É
anime_dance_pivot_filtered = filtered_df.pivot_table(index='Style', columns='Fav_anime', aggfunc='size', fill_value=0)

# –°—Ç—Ä–æ–∏–º —Ç–µ–ø–ª–æ–≤—É—é –∫–∞—Ä—Ç—É
plt.figure(figsize=(16, 8))
sns.heatmap(anime_dance_pivot_filtered, cmap='coolwarm', cbar=True, linewidths=0.2)
plt.title('The relationship between dance style and anime preferences ')
plt.xlabel('Anime')
plt.ylabel('Style')
plt.xticks(rotation=90, ha='right')
plt.yticks(rotation=0)
plt.savefig('anime')
plt.show()
```
